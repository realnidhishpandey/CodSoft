```python
X_custom = data['annual Salary']
y_custom = data['car_purchase_amount']
```

```python
from sklearn.model_selection import train_test_split
X_train_custom, X_test_custom, y_train_custom, y_test_custom = train_test_split(X_custom, y_custom, train_size=0.8,  test_size=0.2, random_state=42)
```

```python
#train dataset view
X_train_custom.head()
```

```python
y_train_custom.head()
```

```python
import statsmodels.api as sm
```

```python
# adding constant to get the intercept
X_train_sm_custom = sm.add_constant(X_train_custom)

# Fit the regression line using 'OLS'
lr_custom = sm.OLS(y_train_custom, X_train_sm_custom).fit()
```

```python
# the intercept and the slope of the regression line fitted
lr_custom.params
```

```python
# performance analysis

print(lr_custom.summary())
```

```python
plt.scatter(X_train_custom, y_train_custom)
plt.plot(X_train_custom, 9154.154204 + 0.561331 * X_train_custom, 'r')
plt.show()
```

```python
y_train_pred_custom = lr_custom.predict(X_train_sm_custom)
res_custom = (y_train_custom - y_train_pred_custom)
```

```python
with sns.axes_style('whitegrid'):
    fig_custom = plt.figure(figsize=(8, 6))
    sns.distplot(res_custom, bins=15, color='skyblue', hist_kws={'edgecolor': 'black', 'linewidth': 1.2})
    fig_custom.suptitle('Distribution of Error Terms', fontsize=18, y=0.95, ha='center', fontweight='bold', color='navy')
    plt.xlabel('Error (y_train_custom - y_train_pred_custom)', fontsize=14, labelpad=12, color='darkslategray')
    plt.ylabel('Density', fontsize=14, labelpad=12, color='darkslategray')

    # Adding a custom background color
    plt.gca().set_facecolor('whitesmoke')

    # Adding a legend
    plt.legend(['Error Terms'], loc='upper right', frameon=True, facecolor='lightgray')

    # Adding grid lines
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Adding a shadow effect to the title
    fig_custom.text(0.5, 0.97, 'Error Analysis', fontsize=20, ha='center', va='center', color='darkblue', fontweight='bold', backgroundcolor='lightyellow')

plt.show()
```

```python
with sns.axes_style('whitegrid'):
    fig_custom_scatter = plt.figure(figsize=(10, 6))

    # Scatter plot
    plt.scatter(X_train_custom, res_custom, color='coral', alpha=0.7, edgecolor='black', linewidth=0.5)

    # Title and labels
    fig_custom_scatter.suptitle('Scatter Plot of Features vs. Error Terms', fontsize=18, y=0.95, ha='center', fontweight='bold', color='darkred')
    plt.xlabel('Features (X_train_custom)', fontsize=14, labelpad=12, color='darkred')
    plt.ylabel('Error Terms', fontsize=14, labelpad=12, color='darkred')

    # Adding a custom background color
    plt.gca().set_facecolor('whitesmoke')

    # Adding grid lines
    plt.grid(axis='both', linestyle='--', alpha=0.7)

    # Adding a legend
    plt.legend(['Scatter Plot'], loc='upper right', frameon=True, facecolor='lightgray')

    # Adding a shadow effect to the title
    fig_custom_scatter.text(0.5, 0.97, 'Error Analysis', fontsize=20, ha='center', va='center', color='darkred', fontweight='bold', backgroundcolor='lightyellow')

plt.show()
```

```python
# Add a constant to X_test_custom
X_test_sm_custom = sm.add_constant(X_test_custom)

# Predict the y values corresponding to X_test_sm_custom
y_pred_custom = lr_custom.predict(X_test_sm_custom)
```

```python
y_pred_custom.head(6)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
np.sqrt(mean_squared_error(y_test_custom, y_pred_custom))
```

```python
r_squared_custom = r2_score(y_test_custom, y_pred_custom)
r_squared_custom
```
